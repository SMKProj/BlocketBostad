```python
res = pd.DataFrame(data, columns=['ID','Published', 'Location', 'Type','Dimension', 'Furnished', 'Accomodate','Description','Features','LandLord'])
res.tail(5)
```

```python
len(res)
```

```python
res['Published'] = res['Published'].str.lstrip('Published')
res['Published'] = res['Published'].str.lstrip('Published')
```

```python
res.head(3)
```

```python
res[['Address', 'City','None']] = res['Location'].str.split(',', expand=True)
```

```python
res[['Rooms', 'Size (m²)']] = res['Dimension'].str.split(',', expand=True)
```

```python
res['Accomodate'] = res['Accomodate'].str.lstrip('Up to')
```

```python
res.head(3)
```

```python
for f in range(len(res['Features'])):
  newString = re.findall('[A-Z][^A-Z]*', res['Features'][f])
  print(newString)
  try:
    kit = newString.index('Kitchen')
    bath = newString.index('Bathroom')
    print(kit,bath)
    res[['Kitchen']] = str(newString[kit : bath])
    res[['Bathroom']] = str(newString[bath+1:] )
  except:
    print('sorry')


```

```python
res.head(5)
```

```python
res['Rooms'] = res['Rooms'].str.rstrip('rooms')
res['Size (m²)'] = res['Size (m²)'].str.rstrip('m²')

res = res.drop(['Location', 'Dimension', 'Features', 'None'], axis=1)
```
